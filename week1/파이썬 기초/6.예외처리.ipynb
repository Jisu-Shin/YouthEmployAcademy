```python
try:
    a=10/0
except ZeroDivisionError:
    print("0으로 나눌수 없습니다.")
```

```python
try:
    a=10/0
except ZeroDivisionError:
    print("0으로 나눌수 없습니다.")
else:
    print("에러 없이 끝났습니다")
finally:
    print("모든 연산이 끝났습니다.")
```

```python
try : 
    a=10/0
except Exception:
    print("에러발생")
```

```python
try:
    a=[3,4,5]
    print(a[3])
except ZeroDivisionError:
    print("zero Division Error 발생")
except IndexError:
    print("Index Error 발생")
except Exception:
    print("Exception")
```

```python
class NotMatchArgsError(Exception):
    pass

def calc(*args):
    if len(args)<2:
        raise NotMatchArgsError
```
