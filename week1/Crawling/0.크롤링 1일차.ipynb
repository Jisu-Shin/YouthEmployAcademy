## 강사님 코드

```python
ul_tags = soup.find('ul',class_='type01')
news_li = ul_tags.find_all('li',recursive=False) # 바로 하위에 있는 자식만 
print(len(news_li))
```

```python
result_list=[]
for news in news_li:
    title_tag = news.find('a',class_='_sp_each_url')
    title = title_tag.text
    url=title_tag.get('href')
    desc = news.find_all('dd')
    desc = desc[1].text
    result_list.append({
        'title' : title,
        'url': url,
        'desc':desc
    })
print(result_list)
```

# 실시간 토픽

- style 은 새로 만들어지는 것이여서 접근 할 수 없없다

```python
url="https://search.naver.com/search.naver"
params = {
    'where':'news',
    'query':'방탄소년단'
}
resp = requests.get(url,params=params)
resp
```

```python
soup = BeautifulSoup(resp.text)
```

```python
ul_tags = soup.find('div',class_='realtime_srch _aside_news_tab')
ol_tags = ul_tags.find_all('ol',recursive=False)
titles=['뉴스','연예.스포츠']
result_dict={}
for title,ol in zip(titles,ol_tags):
    list_li=ol.select('li>a>span')
    rank=[]
    for span in list_li:
        rank.append(span.find('span').text)
    result_dict[title]=rank

result_dict
```

# 블로그 리스트 5페이지까지..

## 강사님 코드

```python
url = 'https://search.naver.com/search.naver'
page_num = int(input("PageNumber?"))
blog_list = []

for i in range(page_num):
    params = {
        'where':'post',
        'query':'방탄소년단',
        'start':i*10+1
    }
    resp = requests.get(url,params=params)
    soup = BeautifulSoup(resp.text)
    blog_list_tags=soup.find_all('li',class_='sh_blog_top')
    
    for blog in blog_list_tags:
        a_tag = blog.find('a',class_='sh_blog_title')
        blog_url=a_tag.get('href')

        title = a_tag.text
        desc = blog.find_all('dd')[1].text

        blog_list.append({
            'url':blog_url,
            'title':title,
            'desc':desc
        })
        

blog_list
```

# 블로그 크롤링 함수

```python
def crawl_blog(query,page_num=2):
    url = 'https://search.naver.com/search.naver'
    blog_list = []
    for i in range(page_num):
        params = {
            'where':'post',
            'query':query,
            'start':i*10+1
        }
        
        resp = requests.get(url,params=params)
        soup = BeautifulSoup(resp.text)
        blog_list_tags=soup.find_all('li',class_='sh_blog_top')
        
        for blog in blog_list_tags:
            a_tag = blog.find('a',class_='sh_blog_title')
            blog_url=a_tag.get('href')

            title = a_tag.text
            desc = blog.find_all('dd')[1].text

            blog_list.append({
                'url':blog_url,
                'title':title,
                'desc':desc
            })
    return blog_list
```

```python
blog_search_list=crawl_blog("2학기",3)
print(len(blog_search_list))
blog_search_list
```

# 뉴스상세페이지
- 뉴스
- 링크,제목,그 안의 내용들까지
[리스트{'title', 'url', 'content'} 안에]

```python
url="https://search.naver.com/search.naver"
params={
    'where':'news',
    'query':'방탄소년단'
}
resp = requests.get(url,params=params)
soup = BeautifulSoup(resp.text)
```

```python
a_tags=soup.select('ul.type01>li>dl>dt>a')
print(len(a_tags))
print(type(a_tags))
```

```python
result_list=[]
for a_tag in a_tags:
    title = a_tag.text
    news_url=a_tag.get('href')
    detail_resp = requests.get(news_url)
    content=detail_resp.text
    result_list.append({
        'title':title,
        'url':news_url,
        'content':content
    })
result_list
```

## 내가 짠 코드

## 블로그 내가 짠 코드

```python
url="https://search.naver.com/search.naver"
params={
    'where':'post',
    'query':'방탄소년단'
}
resp = requests.get(url,params)
soup = BeautifulSoup(resp.text)
ul_tag = soup.find('ul',class_='type01')
li_tags=ul_tag.find_all('li',recursive=False)
news_list=[]
for li in li_tags:
    a=li.find('dt').find('a')
    link=a.get('href')
    resp_link=requests.get(link)
    soup_link=BeautifulSoup(resp_link.text)
    body = soup_link.find('body').find('iframe').get('src')
    news_list.append({
        'title':a.text,
        'url':link,
        'content':body
    })
news_list
```
