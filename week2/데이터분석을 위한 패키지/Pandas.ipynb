```python
print(ktx_df['경부선 KTX'].where(lambda x:x>=40000,1))
ktx_df['경전선 KTX'].where(lambda x:x<5000,0)
```

```python
def categorizing_by_5000(data):
    data=data.where(lambda x:x>=5000,0) #조건의 순서 중요 / false 일때만 바꿈
    data = data.where(lambda x:x<5000,1)
    return data
```

```python
ktx_df.iloc[:,:-2].apply(categorizing_by_5000)
#ktx_df.iloc[:,:-2].apply(categorizing_by_5000,inplace=True)
```
